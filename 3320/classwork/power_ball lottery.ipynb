```python
powerballl = 26
number = 69
dic = {}#number dictionary
pb = {}#powerball dictionary
count = 0
frequency_num = [0 for x in range(0,69)]
frequency_pb = [0 for x in range(0,26)]
file = open('pbnumbers.txt')

for line in file:
    count += 1
    power = line[15:17]
    if power in pb.keys():
        pb[power] = pb.get(power) + 1
    else:
        pb[power] = 1
    for i in range(0,16,3):
        
        if line[i:i+2] in dic.keys():
            dic[line[i:i+2]][0] += 1
            dic[line[i:i+2]][1] = count
        else:
            #次数,出现行
            dic[line[i:i+2]] = [1,count]
            
#根据出现次数排序
least_com = sorted(dic.items(),key=lambda x:x[1],reverse=False)
most_com = sorted(dic.items(),key=lambda x:x[1],reverse=True) 

print('The 10 least common numbers:',end = ' ')
for i in range(0,10):
    print(least_com[i][0],end = ' ')
    
print('\nThe 10 most common numbers:', end =' ')
for i in range(0,10):
    print(most_com[i][0], end = ' ')

#根据最后出现行数排序 
most_overdue = sorted(dic.items(),key=lambda x:x[1][1],reverse=False)

print('\nThe 10 most overdue numbers:', end =' ')
for i in range(0,10):
    print(most_overdue[i][0], end = ' ')

print('\nThe frequency of number 1-69:')
for i in range(1,70):
    if dic.get(str(i)) != None:
        print(f'{i}: {dic[str(i)][0]}')
    else: 
        print(f'{i}: 0')
    
print('\nThe frequency of Powerball number 1-26:')
for i in range(1,27):
    if pb.get(str(i)) != None:
        print(f'{i}: {pb[str(i)]}')
    else: 
        print(f'{i}: 0')

```
