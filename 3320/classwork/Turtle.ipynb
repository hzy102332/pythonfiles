```python
# this program users ask to enter engle and force to hit the target
import turtle
import numpy as np
# constants(the environment of it)

screen_width = 600
screen_hight = 600
target_lleft_x = np.random.uniform(-300,270)
target_lleft_y = np.random.uniform(-300,270)
target_width = 20
force_factor = 30 # a number that will multiply to the distance
projectile_speed = 1
north = 90
south = 270
east = 0
west = 180

# setup the window
turtle.setup(screen_hight, screen_width)

# draw the target
turtle.hideturtle()
turtle.speed(0)
turtle.penup()
turtle.goto(target_lleft_x, target_lleft_y)
turtle.pendown()
turtle.circle(target_width)
turtle.penup()
turtle.goto(0, 0)
turtle.setheading(north)
turtle.showturtle()
turtle.speed(projectile_speed)

totalpoint = 0
goal = 5
while(totalpoint < goal):
    # Get the angle and force from user
    angle = float(input("Enter the projectile's angle(Enter 361 to exit): "))
    force = float(input("enter the launch force: "))
    if angle == 361:
        print("Bye")
        break
    elif angle > 360:
        angle = float(input("You have entered the wrong angle\n Please enter again:"))
    # calculate the distance
    distance = force * force_factor

    # set the heading
    turtle.setheading(angle)

    # launch the projectile
    turtle.pendown()
    turtle.forward(distance)

    # Did it hit the target?
    if(turtle.xcor() >= target_lleft_x -target_width and
        turtle.xcor() <= target_lleft_x + target_width and
        turtle.ycor() >= target_lleft_y - target_width and
        turtle.ycor() <= (target_lleft_y + target_width)):
        print("Target hit!!! You've got 1,000,000 points")
        totalpoint += 1
        if totalpoint < goal:
            target_lleft_x = np.random.uniform(-300,270)
            target_lleft_y = np.random.uniform(-300,270)
            turtle.penup()
            turtle.goto(target_lleft_x, target_lleft_y)
            turtle.pendown()
            turtle.circle(target_width)
            # turtle.setheading(east)
            # turtle.forward(target_width)
            # turtle.setheading(north)
            # turtle.forward(target_width)
            # turtle.setheading(west)
            # turtle.forward(target_width)
            # turtle.setheading(south)
            # turtle.forward(target_width)
            turtle.penup()
            turtle.goto(0, 0)
            turtle.setheading(north)
            turtle.showturtle()
            turtle.speed(projectile_speed)
    else:
        print("You missed the target~    \nTry again !")
    turtle.undo()
    turtle.goto(0, 0)
    turtle.setheading(east)
    turtle.showturtle()
    turtle.speed(projectile_speed)
exit()
turtle.done()
```
