```python
import numpy as np
import jieba
from collections import Counter
import json
import matplotlib.pyplot as plt
import matplotlib
import sys
```

```python
#在引入jieba模块后加入这行代码，代码即可不报错
jieba.setLogLevel(jieba.logging.INFO)

#解决matplotlib显示中文乱码的问题
plt.figure(figsize=(20,20))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.family']='sans-serif'

xianni=open('train_output.txt','r').read()
xianni_words = [x for x in jieba.cut(xianni) if len(x) >= 1 and x != "\n" and x != ":" and x.isdigit() == False] #将全文分割，并将大于两个字的词语放入列表
c=Counter(xianni_words).most_common(10) #取最多的10组
print(json.dumps(c, ensure_ascii=False))

name_list=[x[0] for x in c] #X轴的值
num_list=[x[1] for x in c] #Y轴的值
b=plt.bar(range(len(num_list)), num_list,tick_label=name_list)#画图

plt.xlabel(u'course name')
plt.ylabel(u'times')
# plt.title()
plt.show()#展示
```

```python
filename = "train_output.txt"
infile = open(filename,'r')
dic = {}
for line in infile.readlines():
    line = line.replace("\n","")
    if not line.isdigit() and line != "": #是否包含数字
        if line in dic:
            dic[line] = dic[line] + 1
        else:
            dic[line] = 1

sorted(dic.items(), key = lambda kv:(kv[1], kv[0]))
# sorted(dic)
```
