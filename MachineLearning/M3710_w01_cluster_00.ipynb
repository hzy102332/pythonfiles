```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# construct K-means algorithm
K=np.arange(1,11)
Dist=[]
Labels=[]
Centroids = []
for k in K:
    kmeans3 = KMeans(init="random", n_clusters=k,n_init=10,max_iter=300,
                random_state=42)
    kmeans3 = kmeans3.fit(Data)
    # kmeans = KMeans(Data=Data)
    #Labels, centroids, dist = kmeans3.cluster(k,None)
    # Labels, centroids, dist = kmeans3.fit(k,None)
    # centroid # 查看质心
    center = kmeans3.cluster_centers_
    Centroids.append(center)
    # inertia
    dist = kmeans3.inertia_
    Dist.append(dist)
    # y_pred
    y_pred = kmeans3.labels_
    Labels.append(y_pred)
    
    # Dist.append(dist)
    # Labels.apend(Label)
    # Centroids.append(centroids)
    print("the centroid under k: " %(k))
    print(Centroids)
plt.plot(K,Dist)
plt.grid(True)
plt.xlabel("k")
plt.ylabel("Inertia Distance after clustering")
#plt.close()
plt.show()
```
