```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
x,y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)
```

```python
color = ['red','pink','orange','gray']
fig, axi1 = plt.subplots(1)
for i in range (4):
    axi1.scatter(x[y==i,0],x[y==i,1],marker='o',s =8,c = color[i])
plt.show()
```

```python
from sklearn.cluster import KMeans
n_cluster = 3
cluster = KMeans(n_clusters = n_cluster,random_state=0).fit(x)
 # prediction
y_pred = cluster.labels_
y_pred
```

```python
centroid = cluster.cluster_centers_
centroid
```

```python
inertia = cluster.inertia_
inertia
```

```python
y_pred = cluster.labels_
y_pred
```

```python
color = ['red','pink','orange','green']
fig,axi1 = plt.subplots(1)
for i in range(3):
    axi1.scatter(x[y_pred==i,0],x[y_pred==i,1],marker='o',s = 8,c= color[i])
# centroid
axi1.scatter(centroid[:,0],centroid[:,1],marker='X',s=100,c='green')
```
